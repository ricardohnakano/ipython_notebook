```python
#https://code.google.com/p/bigquery-e2e/source/browse/samples/ch06/run_load.py
from apiclient.http import MediaFileUpload
import time
import json
import datetime

import auth
import run_load
```

```python
service = auth.build_bq_client()
```

```python
# Delete a table
def deleteTable(projectId, datasetId, tableId):
    service.tables().delete(projectId=projectId, datasetId=datasetId, tableId=tableId).execute()
```

```python
def loadTable(projectId,datasetId, load_file, table):

    # Load configuration with the destination specified.
    load_config = {
      'destinationTable': {
        'projectId': projectId,
        'datasetId': datasetId,
        # You can update this for each example.
        'tableId': table
      }
    }
    # Setup the job here.
    # load[property] = value
    load_config['schema'] = {
      'fields': [
        {'name':'id', 'type':'integer'},
        {'name':'tc_id', 'type':'integer'},
        {'name':'trip_id', 'type':'string'},
        {'name':'main_leg', 'type':'integer'},
        {'name':'trip_link', 'type':'string'},
        {'name':'driver_id', 'type':'integer'},
        {'name':'trip_created_at', 'type':'timestamp'},
        {'name':'trip_departure_datetime', 'type':'timestamp'},
        {'name':'departure_address', 'type':'string'},
        {'name':'destination_address', 'type':'string'},
        {'name':'departure_city', 'type':'string'},
        {'name':'destination_city', 'type':'string'},
        {'name':'departure_state', 'type':'string'},
        {'name':'destination_state', 'type':'string'},
        {'name':'departure_country', 'type':'string'},
        {'name':'destination_country', 'type':'string'},
        {'name':'country', 'type':'string'},
        {'name':'seats_offered', 'type':'integer'},
        {'name':'trip_distance', 'type':'float'},
        {'name':'suggested_price', 'type':'integer'},
        {'name':'price', 'type':'integer'},
        {'name':'trip_canceled', 'type':'integer'},
        {'name':'auto_accept', 'type':'integer'},
        {'name':'trip_comment', 'type':'integer'},
        {'name':'driver_message', 'type':'integer'},
        {'name':'pax_message', 'type':'integer'},
        {'name':'recurrent', 'type':'integer'},
        {'name':'ladies_only', 'type':'integer'},
        {'name':'is_return', 'type':'integer'},
        {'name':'seats_booked', 'type':'integer'},
        {'name':'seats_approved', 'type':'integer'},
        {'name':'seats_canceled', 'type':'integer'},
        {'name':'seats_rejected', 'type':'integer'},
        {'name':'promo_code', 'type':'string'},
        {'name':'description_code', 'type':'string'}
      ]
    }

    upload = MediaFileUpload(load_file, mimetype='application/octet-stream', resumable=True)
    # End of job configuration
    run_load.start_and_wait(service.jobs(),
                            'kinetic-valor-110522',
                            load_config,
                            media_body=upload)
```

```python
def main():
    inicio = datetime.datetime.now() 
    deleteTable('kinetic-valor-110522', 'dataset', 'trips')
    loadTable('kinetic-valor-110522','dataset', 'trips6.csv','trips')
    print datetime.datetime.now() - inicio
```

```python
if __name__ == '__main__':
    main()
```
