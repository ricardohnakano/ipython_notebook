```python
distance_reta = []
for row in range(0,len(distance.index)):
        dep = (distance.xs(row)[0],distance.xs(row)[1])
        dest = (distance.xs(row)[2],distance.xs(row)[3]) 
        distance_reta.append(vincenty(dep, dest).kilometers)
```

```python
distance['line_distance'] = distance_reta
```
